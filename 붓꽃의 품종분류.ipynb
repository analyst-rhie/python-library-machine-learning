### 1.7 붓꽃의 품종 분류 (p.39)
---

* 우리의 목표 : 어떤 품종인지 구분해놓은 측정 데이터를 이용해 새로 채집한 붓꽃의 품종을 예측하는 머신러닝 모델을 만드는 것.   
 * 문제의 종류 : 분류문제   
 * 클래스(class) : 붓꽃의 종류 (출력될 수 있는 값)   
 * 레이블(label) : 데이터 포인트 하나(붓꽃 하나)에 대한 기대 출력은 꽃의 품종이므로 꽃의 품종이 레이블   
* 용어 정리
 * 각 행(아이템) : 샘플
 * 속성 : 특성 
 * data 배열의 크기 : 샘플의수에 특성의 수를 곱한 값으로 이는 scikit-learn의 스타일이며 항상 데이터가 이런구조일 것이라 가정
* 코드 규칙
 * scikit - learn 에서 **데이터**는 **대문자 X**, **레이블**은 **소문자 y**로 표기.
  * 이는 수학에서 함수의 입력을 x, 출력을 y로하는 표준공식에서 유래되며, 데이터는 2차원 배열(행렬)이므로 대문자 X, 타깃은 1차원 배열(벡터)이므로 소문자 y 사용
---

```python
# 데이터적재
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity="all"
from sklearn.datasets import load_iris
iris_dataset = load_iris()
```

```python
# iris 객체는 파이썬의 딕셔너리와 유사한 Bunch 클래스의 객체, 즉 키와 값으로 구성 됨.
iris_dataset.keys()
```

```python
# 데이터 셋 크기
iris_dataset['data'].shape
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(iris_dataset['data'], iris_dataset['target'], random_state = 0)
# 여기서 iris_dataset 은 데이터를 골고루 섞어주기 위함.
# X_train은 전체 데이터 세트의 75%를, X_test는 나머지 25%를 담고 있음.
```

```python
# 특성의 수가 적을 때 모든 특성을 짝지어 만드는 산점도 행렬(scatter matrix) 사용이 좋다.
import pandas as pd
import mglearn
iris_dataframe = pd.DataFrame(X_train, columns = iris_dataset.feature_names)
# 데이터프레임을 사용해 y_train에 따라 색으로 구분된 산점도 행렬을 만듦.
pd.plotting.scatter_matrix(iris_dataframe, c=y_train)#, figsize=(15, 15), marker='o', hist_kwds={'bins': 20}, s=60, alpha=.8, cmap=mglearn.cm3)
```
